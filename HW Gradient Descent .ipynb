```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

$定義函數 y = 20xsin(x) + x^2 + x^3$

```python
def Lossf(x):
    Lf = 20*x*np.sin(x) + x**2 + x**3
    return Lf
def diffLossf(x):
    diffLf = 20*x*np.cos(x) + 2*x + 3*(x**2)
    return diffLf
```

```python
X = np.linspace(-10,10,100)
plt.plot(X, Lossf(X))
```

$定義梯度下降函數$

```python
def gradientFunction(x_start, learning_rate, iteration):
    xwalk = np.zeros(iteration + 1)
    xwalk[0] = x_start
    for i in range (iteration):
        x_new = xwalk[i] - (learning_rate)*(diffLossf(xwalk[i]))
        xwalk [i+1] = x_new
    return xwalk
```

$初始值從9開始，進行迭代10000次$

```python
x = gradientFunction(9, 0.05, 10000)
t = np.linspace(-10,10,100)
plt.figure(figsize=(10,10))
plt.plot(t, Lossf(t))
plt.scatter(x, Lossf(x), c=x)
print('It just find local minimum, x= ',float(x[10000]), 'Lossfunction value =', float(Lossf(x[10000])))
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib as mpl
fig = plt.figure()
ax = fig.gca(projection='3d')
figure = ax.plot(t,Lossf(t),zdir='y', c='b')
figure = ax.bar(x,Lossf(x),zdir='y', color='r')
plt.show()
```
